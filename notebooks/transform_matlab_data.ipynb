# Load and Plot

Now lets load that data into a dataframe and plot it

```python
data_path = r'C:\Users\diego\OneDrive\Documentos\PENDULO INVERTIDO\Pendulo Invertido Diego\Python-Furuta-Pendulum\data\datasets\synthetic_data2.parquet'
data = pd.read_parquet(data_path)
```

```python
data.shape
```

```python
def plot_data(data: pd.DataFrame, title: str, init: int = 0, end:int = 100) -> None:
    fig, axs = plt.subplots(nrows=2, ncols=4, figsize=(10, 10))
    axs = axs.flatten()

    for idx, column in enumerate(data.columns):
        axs[idx].plot(data[column][init:end])
        axs[idx].set_title(column)

    fig.subplots_adjust(hspace=0.4, wspace=0.6)
    fig.suptitle(title)
```

```python
plot_data(data=data,
          title='Square Data',
          init=2000,
          end=12000)
```
